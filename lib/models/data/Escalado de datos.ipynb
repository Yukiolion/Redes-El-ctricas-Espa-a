```python
import pandas as pd
import numpy as np
import pickle


from sklearn.preprocessing import MinMaxScaler

```

```python
df = pd.read_csv("demanda-depurado.csv")

df
```

```python
df['fecha'] = pd.to_datetime(df['fecha'])
df['dia'] = df['fecha'].dt.day
df['mes'] = df['fecha'].dt.month
df['año'] = df['fecha'].dt.year

# Selección de features y target
X = df[['dia', 'mes', 'año']].values
y = df[['valor']].values

# Escalar datos
X_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()

X_scaled = X_scaler.fit_transform(X)
y_scaled = y_scaler.fit_transform(y)

# Crear secuencias temporales
T = 7
X_seq = []
y_seq = []

for i in range(len(X_scaled) - T):
    X_seq.append(X_scaled[i:i+T])
    y_seq.append(y_scaled[i+T])

X_seq = np.array(X_seq)
y_seq = np.array(y_seq)

# Usamos secuencias temporales porque el orden es importante y no se puede mezclar aleatoriamente como en train_test_split
secuencia_temp = int(len(X_seq) * 0.7)
X_train_seq = X_seq[:secuencia_temp]
y_train_seq = y_seq[:secuencia_temp]
X_test_seq = X_seq[secuencia_temp:]
y_test_seq = y_seq[secuencia_temp:]
```

```python
with open("X_scaler.pkl", "wb") as file:
    pickle.dump(X_scaler, file)

with open("y_scaler.pkl", "wb") as file:
    pickle.dump(y_scaler, file)
```
